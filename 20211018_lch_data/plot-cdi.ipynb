```python
import pandas as pd
import numpy as np

df = pd.read_csv('data.csv')
theta1_raw = df['theta2'].to_list()
theta4_raw = df['theta3'].to_list()
theta3_raw = df['theta1'].to_list()
theta2_raw = df['theta4'].to_list()

theta1 = (theta1_raw - np.ones(len(theta1_raw))*theta1_raw[0])
theta2 = (theta2_raw - np.ones(len(theta2_raw))*theta2_raw[0])
theta3 = (theta3_raw - np.ones(len(theta3_raw))*theta3_raw[0])
theta4 = (theta4_raw - np.ones(len(theta4_raw))*theta4_raw[0])

import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(15, 10))
plt.plot(theta1, label="theta1")
plt.plot(theta2, label="theta2")
plt.plot(theta3, label="theta3")
plt.plot(theta4, label="theta4")
plt.legend()
```

```python
r = 0.2
l = 0.3
w = 0.3

alpha = 1/(l+w)

F = np.array([ [-alpha, alpha, alpha, -alpha],
               [1, 1, 1, 1],
               [-1, 1, -1, 1] ])*r/4
```

```python
theta = np.array( [theta1, theta2, theta3, theta4] )
```

```python
twist = F@theta
twist
```

```python
wbz = twist[0]
vbx = twist[1]
vby = twist[2]
```

```python
plt.figure(figsize=(15, 15))
plt.plot(vbx,vby)
plt.figure(figsize=(15, 15))
plt.plot(wbz)
```

```python
x = (vbx*np.sin(wbz) + (np.cos(wbz)-1)*vby)/wbz
y = (vby*np.sin(wbz) + (1-np.cos(wbz))*vbx)/wbz

Tsb = np.array( [ [np.cos(wbz), -np.sin(wbz), 0, x],
                  [np.sin(wbz), np.cos(wbz), 0, y],
                  [0,0,1,0],
                  [0,0,0,1] ] )
```

```python
plt.figure(figsize=(15, 15))
plt.plot(x,y)
```
